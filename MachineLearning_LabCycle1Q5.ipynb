```python
import pandas as pd

# Manually creating the dataset
data = [
    ['milk', 'bread', 'butter'],
    ['milk', 'bread'],
    ['bread', 'butter'],
    ['milk', 'bread', 'butter'],
    ['bread', 'butter'],
    ['milk', 'butter'],
    ['milk', 'bread', 'butter'],
    ['bread', 'butter'],
    ['milk', 'bread']
]

# Converting to a pandas DataFrame for easier manipulation
df = pd.DataFrame(data, columns=['item1', 'item2', 'item3'])
df = df.apply(lambda x: x.dropna().tolist(), axis=1)

# Show the dataset
df.head()
```

```python
# Display the first few transactions
print("First few transactions:")
print(df.head())

# Check the total number of transactions
print(f"\nTotal number of transactions: {len(df)}")

# Check the unique items in the dataset
all_items = [item for sublist in df for item in sublist]
unique_items = set(all_items)
print(f"\nTotal unique items: {len(unique_items)}")
print(f"Unique items: {unique_items}")
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Convert dataframe to list of lists
transactions = df.tolist()

# Apply TransactionEncoder
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

# Create new dataframe
encoded_df = pd.DataFrame(te_array, columns=te.columns_)

print(encoded_df.head())
```

```python
# Calculate support for each individual item
support = encoded_df.mean()

print("Support for each item:\n")
print(support)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from mlxtend.frequent_patterns import apriori

# Apply Apriori
frequent_itemsets = apriori(encoded_df,
                            min_support=0.5,
                            use_colnames=True)

print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules
rules = association_rules(frequent_itemsets,
                          metric="confidence",
                          min_threshold=0.7)

# Display rules
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
#The generated association rules show high confidence values (> 0.75), indicating that items such as butter and milk are frequently purchased with bread. However, the lift values are close to 1, suggesting that the associations are not particularly strong and may be influenced by the high overall frequency of bread purchases.
```
