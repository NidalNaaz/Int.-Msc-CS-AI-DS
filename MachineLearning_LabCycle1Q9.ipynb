```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.linear_model import LogisticRegression
```

```python
iris = load_iris()

X = iris.data
y = iris.target

print("Dataset shape:", X.shape)
```

```python
def evaluate_model(X_transformed, y):

    model = LogisticRegression(max_iter=1000)
    scores = cross_val_score(model, X_transformed, y, cv=5)

    print("Accuracy:", scores)
    print("Mean Accuracy:", scores.mean())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("PCA shape:", X_pca.shape)

evaluate_model(X_pca, y)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)

X_lda = lda.fit_transform(X, y)

print("LDA shape:", X_lda.shape)

evaluate_model(X_lda, y)
```

```python
tsne = TSNE(n_components=2, random_state=42)

X_tsne = tsne.fit_transform(X)

print("TSNE shape:", X_tsne.shape)

evaluate_model(X_tsne, y)
```

```python
svd = TruncatedSVD(n_components=2)

X_svd = svd.fit_transform(X_scaled)

print("SVD shape:", X_svd.shape)

evaluate_model(X_svd, y)
```

```python
pca3 = PCA(n_components=3)

X_pca3 = pca3.fit_transform(X_scaled)

print("PCA (3) shape:", X_pca3.shape)

evaluate_model(X_pca3, y)
```

```python
tsne3 = TSNE(n_components=3, random_state=42)

X_tsne3 = tsne3.fit_transform(X)

print("TSNE (3) shape:", X_tsne3.shape)

evaluate_model(X_tsne3, y)
```

```python
svd3 = TruncatedSVD(n_components=3)

X_svd3 = svd3.fit_transform(X_scaled)

print("SVD (3) shape:", X_svd3.shape)

evaluate_model(X_svd3, y)
```
