```python
# Numerical computations
import numpy as np

# For visualization (optional but good for viva)
import matplotlib.pyplot as plt

# MNIST dataset + preprocessing utilities
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Splitting dataset
from sklearn.model_selection import train_test_split
```

```python
# Load MNIST dataset
(X_train_full, y_train_full), (X_test, y_test) = mnist.load_data()

# Check shapes
print("Training data shape:", X_train_full.shape)
print("Training labels shape:", y_train_full.shape)
print("Test data shape:", X_test.shape)
print("Test labels shape:", y_test.shape)
```

```python
# Convert pixel values from int (0–255) to float (0–1)
X_train_full = X_train_full.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
```

```python
# Convert class labels to one-hot encoded vectors
y_train_full_encoded = to_categorical(y_train_full, num_classes=10)
y_test_encoded = to_categorical(y_test, num_classes=10)

# Check one example
print("Original label:", y_train_full[0])
print("One-hot encoded:", y_train_full_encoded[0])
```

```python
# Split training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full,
    y_train_full_encoded,
    test_size=0.2,
    random_state=42
)

# Check final shapes
print("Final Training shape:", X_train.shape)
print("Validation shape:", X_val.shape)
print("Test shape:", X_test.shape)
```

```python
# Display first 5 training images
plt.figure(figsize=(10, 2))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(X_train[i], cmap="gray")
    plt.title(f"Label: {np.argmax(y_train[i])}")
    plt.axis("off")
plt.show()
```
