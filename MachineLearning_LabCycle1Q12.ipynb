```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)

kmeans.fit(X)

labels_kmeans = kmeans.labels_

print("K-Means Cluster Labels:")
print(labels_kmeans)
```

```python
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```

```python
kmedoids = KMedoids(n_clusters=3, random_state=42)

kmedoids.fit(X)

labels_kmedoids = kmedoids.labels_

print("K-Medoids Cluster Labels:")
print(labels_kmedoids)
```

```python
print("Medoid Points:")
print(kmedoids.cluster_centers_)
```

```python
X_transposed = X.T

cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    X_transposed,
    c=3,
    m=2,
    error=0.005,
    maxiter=1000
)

labels_fcm = np.argmax(u, axis=0)

print("Fuzzy C-Means Labels:")
print(labels_fcm)
```

```python
print("Cluster Centers:")
print(cntr)
```
